**EMBEDDING**
    
    *  Embedding is the process of converting text into a fixed-size numerical vector.
    *  That preserves the meaning of the text so that machines can understand, compare, and analyze language.
    *  The model text-embedding-004 was trained with an internal representation size = 768.


**Why embedding ?**

    *   Before giving text to any ML / DL model, we must convert it into numbers.
    *   That conversion of words / sentences into vectors is embedding.
    *   Semantic search(used for retriving past history in chatgpt)
    *   similarity(cosine similarity output:ranges from [-1 to 1])

```python
#cosine similarity
import numpy as np

def cosine_similarity(vec1,vec2):
  return np.dot(vec1,vec2)/(np.linalg.norm(vec1)*np.linalg.norm(vec2))

#input text
text1= "Ai is transforming the world"
text2="Artificial Intelligence is changing industries"

#embedding text to vector using text-embedding-004
emb1=client.models.embed_content(
    model="text-embedding-004",
    contents=text1
)

emb2=client.models.embed_content(
    model="text-embedding-004",
    contents=text2
)

#converting the embedded vector to array
emb1=np.array(emb1.embeddings[0].values)
emb2=np.array(emb2.embeddings[0].values)

#similarity calculation
similarity = cosine_similarity(emb1,emb2)
print("Similarity : ",similarity)
```

```python
#semantic search
sentences =[
    "I love ml",
    "the weather is very hot today",
    "ai model uses neural networks",
    "i enjoy studying artificial intelligence"
  ]

query ="Learning Ai"

query_emb = client.models.embed_content(
    model = "text-embedding-004",
    contents = query
)

scores =[]

for sent in sentences:
  sent_emb = client.models.embed_content(
      model = "text-embedding-004",
      contents = sent
  )

  score = cosine_similarity(query_emb.embeddings[0].values,sent_emb.embeddings[0].values)
  scores.append((sent,score))

scores.sort(key=lambda x:x[1],reverse=True)

for sent,score in scores:
  print(sent,"->",score)
```
